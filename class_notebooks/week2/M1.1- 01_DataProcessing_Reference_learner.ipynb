### Key Terms

|   Term | Definition         | 
|:-------|:--------------|
| Variable | A named reference to a value that can be changed. Variables are used to store data in a program and refer to it by a name.|
| Function | A named block of reusable code that performs a specific task. Functions are used to modularize code, making it easier to read, write, and maintain Python programs.|
| In-built functions | Pre-defined functions that are built into the Python language and can be used directly without having to define or install them first: https://docs.python.org/3/library/functions.html.|
| Collections |  Built-in Python data structures that allow us to store and manipulate data such as, lists, tuples, sets, and dictionaries.|
| Pythonic loop |  A construct in Python used to iterate over code blocks to avoid repetitive code and thus write clean, readable, and efficient code (e.g. `for` or `while` loop).|
| Libraries	| Collections of pre-written code that contain pre-written functions etc. providing a set of functionalities and tools to simplify the programming process. |

### Key Python Statements

|   Statement | Definition         | Example      |
|-------:|:--------------|:----------|
|      `type` | A built-in function that allows you to determine the type of a given object. | `type(10)` *# outputs int (short for integer)* | 
|      `str` | A built-in data type called "string" that represents a sequence of characters enclosed within single, double or triple quotes. | `my_string = "Hello, World!"` | 
| `f-string` | A type of string formatting feature where putting the letter "f" before the opening quote of the string allows you to embed expressions inside string literals. | `f"string text {expression} more string text"` |
|     `int` | A built-in data type that represents integers i.e. whole numbers without a fractional component. | `my_int = 42` | 
|     `float` | A built-in data type that represents floating-point numbers i.e. real numbers with a fractional component. | `my_float = 3.14` | 
|     `bool` | A built-in data type that represents boolean values, which can be either `True` or `False`. Boolean values can be assigned to variables directly, or can be the result of logical operations or comparisons. | `10 > 5` *# outputs true* | 
|      `list` | A built-in data structure used to store a mutable, ordered sequence of items separated by commas and enclosed in square brackets, `[ ]`. | `my_list = [1,2,3]` | 
|      `tuple` | A built-in data structure used to store an immutable, ordered sequence of items separated by commas and enclosed in round brackets, `( )`. | `my_tuple = (1,2,3)` | 
|      `set` | A built-in data structure used to store a mutable, unordered collection of unique items separated by commas and enclosed in curly brackets, `{ }`. | `my_set = {1,2,3}` | 
|      `dict` | A built-in data structure called a "dictionary" used to store a mutable, unordered collection of key-value pairs separated by commas (with each key and their values separated by a colon) and enclosed in curly brackets, `{ }`. | `my_dict = {'name': 'John', 'age': 30, 'city': 'London'}` | 
|     `len` | A built-in function that returns the length or number of items in a collection such as a string, list, tuple, dictionary, or set. | `len("Hello, World!")` *# outputs 13* | 
|     `print` | A built-in function that is used to output or display text on the console or terminal. | `print("Hello, World!")` | 
|     `for` | A control flow statement used as a loop contruct to iterate over a sequence or iterable object such as a list, tuple, string, or dictionary. | `for x in my_list: print(x)` | 
|     `range` | A built-in function that generates a sequence of numbers, which is commonly used with loops. | `range(start, stop, step)` | 
|     `while` | A control flow statement used as a loop contruct to execute a block of code repeatedly as long as a certain condition is true. | `while num < 5: print(num)` | 
| `numpy` | An open-source library that provides a multidimensional array object and an extensive library of mathematical functions for working with arrays. | `import numpy as np` |
|     `array` | Numpy arrays are a type of data structure that provide efficient computation of multi-dimensional arrays and matrices. | `np.array([1,2,3])` |
|     `Pandas` | An open-source data analysis library that provides easy-to-use data structures and data manipulation tools for efficient data analysis and cleaning. | `import pandas as pd` | 
|      `DataFrame` |  A two-dimensional labeled data structure from Pandas that is used to store and manipulate data in a tabular format. | `df = pd.DataFrame({'Name':['Jane','Joe'],'Age':['31','45']})` |
