```python
df = pd.DataFrame({
    "order_id": [101, 102, 103, 104, 105, 106, 107, 108],
    "region": [
        "London", "London", "Manchester", "London",
        "Birmingham", "Manchester", "London", "Birmingham"
    ],
    "category": [
        "Food", "Tech", "Food", "Food",
        "Tech", "Food", "Tech", None
    ],
    "sales": [1200, 800, 400, 1500, 2000, 300, 500, 900],
    "cost":  [700, 900, 300, 600, 1200, 350, 600, 400],
})
```

### Action: filter rows where region == 'London'

```python
df[df['region'] == 'London']
```

### Select all rows where cost is greater than 500

```python
df[df['cost'] > 500]
```

### Select all rows where cost is greater than 600 and region is London

```python
df[(df['cost'] > 600) & (df['region'] == 'London')]
```

### Select all rows where region is either London, Manchester or France

```python
df[df['region'].isin(['London', 'Manchester', 'France'])]
```

### Select all rows where category is missing

```python
df[df['category'].isna()]
```

### Select all rows where category contains 'ec' (case insensitive)

```python
df[df['category'].str.contains('ec', case=False)]
```

# Aggregating Data With Pandas

---

```python
# Data question: What does our raw transactional dataset look like before any summarisation?
import pandas as pd

df = pd.DataFrame(
    {
        "order_id":  [1001,1002,1003,1004,1005,1006,1007,1008,1009,1010],
        "date":      pd.to_datetime(["2026-01-02","2026-01-02","2026-01-03","2026-01-03","2026-01-04",
                                     "2026-01-04","2026-01-05","2026-01-05","2026-01-06","2026-01-06"]),
        "customer":  ["C001","C002","C001","C003","C004","C002","C005","C001","C003","C006"],
        "city":      ["London","London","London","Manchester","Manchester","London","Birmingham","London","Manchester","Birmingham"],
        "category":  ["Coffee","Coffee","Tea","Coffee","Tea","Tea","Coffee","Coffee","Tea","Coffee"],
        "qty":       [2,1,3,1,2,1,4,1,2,1],
        "unit_price":[3.50,3.50,2.40,3.50,2.40,2.40,3.50,3.50,2.40,3.50],
    }
)

df["revenue"] = df["qty"] * df["unit_price"]
df
```

### How many orders do we have in each city?

```python
df.groupby('city')['order_id'].count()
```

### What is the total revenue generated in each city?

```python
df.groupby('city')['revenue'].sum()
```

### What is the average order revenue in each city?

```python
df.groupby('city')['revenue'].mean()
```

### What is the smallest (minimum) order revenue we see in each city?

```python
df.groupby('city')['revenue'].min()
```

### What is the largest (maximum) order revenue we see in each city?

```python
df.groupby('city')['revenue'].max()
```

### What is the typical order revenue in each city (less sensitive to outliers than the mean)?

```python
df.groupby('city')['revenue'].median()
```

### How consistent are order revenues in each city (i.e., how much do they vary)?

```python
df.groupby('city')['revenue'].std()
```

# Introducing `.agg()` — combining multiple aggregates

### What does `.agg()` do?

The `.agg()` method allows us to:

- Apply **multiple aggregate functions at once**
- Give **clear names** to each aggregated output
- Produce **one tidy summary table** instead of many separate ones



```python
# .agg()
df.groupby('city').agg(order=('order_id', 'count'),total_revenue=('revenue', 'sum'))
```

```python
# .agg() multiple metrics
df.groupby('city').agg(
    avg_order=('revenue', 'mean'),
    median=('revenue', 'median')
).reset_index()
```

## Why do we need functions?

Before introducing functions, it’s important to understand **the problem they are solving**.

In real data work — especially during **data preprocessing** — you will often:

- Clean **multiple columns in the same way**
- Copy and paste **working code** to move faster
- Make **small tweaks** as requirements evolve

At first, this approach feels efficient.

---

### The hidden problem

As your dataset grows and your logic changes, this pattern quietly introduces risks:

- Repeated logic becomes **hard to update**
- Small changes can lead to **inconsistent behaviour**
- Bugs are more likely to appear in **only some places**

---

### Why this example is written this way

This example is written for you as a **data analyst transitioning toward data science**.

It deliberately **starts by repeating code**, so you can:

- Clearly see the **pain points**
- Feel why the approach stops scaling
- Understand **why functions exist**, before learning how to write them




### An simulated example

```python
import pandas as pd
import numpy as np

# Simulated raw data
df = pd.DataFrame({
    "city_a": [" London ", "PARIS", None, "  berlin"],
    "city_b": [" madrid", None, "ROME ", " Lisbon"],
    "city_c": [None, " ATHENS", "vienna ", "  Prague"]
})

df
```

```python
# Clean city_a
df["city_a"] = (
    df["city_a"]
    .fillna("unknown")
    .str.strip()
    .str.lower()
)

# Clean city_b
df["city_b"] = (
    df["city_b"]
    .fillna("unknown")
    .str.strip()
    .str.lower()
)

# Clean city_c
df["city_c"] = (
    df["city_c"]
    .fillna("unknown")
    .str.strip()
    .str.lower()
)

df
```

#### We can create a custom function to help us streamline the workflow

```python
def clean_text_column(series: pd.Series) -> pd.Series:
    return (
        series
        .fillna("unknown")
        .str.strip()
        .str.lower()
    )
```

#### Mixing with pythonic loops to make script cleaner

```python
for col in ["city_a", "city_b", "city_c"]:
    df[col] = clean_text_column(df[col])

df
```

# Why does this matter?

- Single source of truth for preprocessing logic

- Easier experimentation (change once, apply everywhere)

- Functions compose cleanly into pipelines later (e.g. sklearn)

- Clear boundary between what you do and where you do it

- This is the first step toward reproducible, testable preprocessing pipelines.

## Function Syntax

```python
def function_name(parameter1, parameter2):
    """Optional description of what the function does."""
    # function body (indented)
    result = parameter1 + parameter2
    return result
```

## Python Function Anatomy

**def**  
- Short for *define*  
- Tells Python you are creating a function  

**function_name**  
- How you refer to and call the function later  
- Should describe what the function does  

**(parameter1, parameter2)**  
- Parameters are inputs to the function  
- Both inputs must be provided when calling the function  

**:**  
- Marks the start of the function body  

**Indentation**  
- Everything indented underneath belongs to the function  
- This is how Python knows what code is inside the function  

**return**  
- Sends a value back to where the function was called  
- Ends the function immediately  


## Function Examples

### 1. A function can take one input and returns one output

```python

def double(number):
    return number * 2

double(4)
```

### 2. The same function can be reused with different values

```python

def add_exclamation(text):
    return text + "!"

add_exclamation("hello")
add_exclamation("great")
add_exclamation("wow")
```

### 3. A function can accept more than one input


```python

def full_name(first, last):
    return first + " " + last

full_name("Ada", "Lovelace")
```

### 4. I can set parameters to have default values


```python
def greet(name, greeting="Hello"):
    return greeting + ", " + name

greet("Sam")
greet("Sam", greeting="Hi")
```

#### 5. Interestingly, without `return`, the function returns `None`

```python
def announce(message):
    print(message)

result=announce("Starting program")
print(result)
```
