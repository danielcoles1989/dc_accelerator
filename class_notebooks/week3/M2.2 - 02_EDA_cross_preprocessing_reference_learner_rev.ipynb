## Key Terms: Train/Test Split, Scaling & Encoding

| Key Term | Definition |
|---|---|
| **Train-Test Split** | Dividing data into training (fit model) and testing (evaluate model) subsets to assess generalisation. |
| **Stratification** | Ensuring train and test sets have representative proportions of each class, especially for imbalanced datasets. |
| **Data Leakage** | When information from the test set or future data inadvertently influences the training process, causing inflated performance metrics. |
| **Feature Engineering** | Creating new features or transforming existing ones before model training to improve predictive power. |
| **Standardisation (Scaling)** | Transforming features to have mean ≈ 0 and standard deviation ≈ 1 using StandardScaler. |
| **One-Hot Encoding** | Converting categorical variables into binary columns (0/1) to represent each category without imposing false ordering. |
| **Ordinal Encoding** | Mapping categorical variables with natural ranking (e.g., Low=0, Medium=1, High=2) to numeric values. |
| **Label Encoding** | Assigning numeric labels to categories; risks introducing false ordering for nominal variables. |
| **Log Transformation** | Applying log function to skewed features to normalise distribution and reduce outlier influence. |
| **Fit vs. Transform** | Fit: learn parameters from training data. Transform: apply those parameters to new data without relearning. |


## Key Python Statements

| Statement | Definition | Example |
|---|---|---|
| `train_test_split()` | Splits data into training and testing subsets with configurable size and stratification. | `X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)` |
| `StandardScaler()` | Transforms features to mean=0 and std=1; must fit on training data only. | `scaler = StandardScaler(); X_train_scaled = scaler.fit_transform(X_train); X_test_scaled = scaler.transform(X_test)` |
| `get_dummies()` | Converts categorical columns to one-hot encoded binary columns. | `X_encoded = pd.get_dummies(X, columns=['city'], drop_first=True)` |
| `LabelEncoder()` | Maps categorical values to numeric labels (0, 1, 2, ...); use with caution for nominal data. | `encoder = LabelEncoder(); X['category'] = encoder.fit_transform(X['category'])` |
| `np.log1p()` | Applies log transformation (log(1+x)) to handle zeros and skewed distributions. | `X['income'] = np.log1p(X['income'])` |
| `KBinsDiscretizer()` | Bins continuous features into discrete categories using uniform or quantile strategy. | `binning = KBinsDiscretizer(n_bins=5, encode='ordinal', strategy='quantile'); X_binned = binning.fit_transform(X_train)` |
| `.corr()` | Calculates Pearson correlation coefficients between numeric columns; useful for detecting multicollinearity. | `correlation_matrix = X_train.corr(); print(correlation_matrix['income'].sort_values(ascending=False))` |
| `.describe()` | Generates summary statistics (mean, std, min, max, quartiles) for numeric columns. | `X_train.describe()` |
| `.drop()` | Removes specified columns or rows; use `axis=1` for columns. | `X_clean = X.drop(['constant_col', 'duplicate_col'], axis=1)` |
| `.fillna()` | Handles missing values by filling with a specified value or strategy (mean, forward fill, etc.). | `X['age'] = X['age'].fillna(X['age'].median())` |
