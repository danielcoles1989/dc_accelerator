## How is `sklearn` structured?

#### What are estimators, including reference examples? 

- An *estimator* is a sklearn object that is capable of inferring some properties on new data. They are used in a wide range of machine learning tasks, such as data preprocessing, clustering, regression, classification and dimensionality reduction.



- The estimation process is performed by the `fit()` method, which either takes only a training dataset as a parameter, or additionally the target for supervised learning algorithms (any other parameter required to guide the estimation is considered a "hyperparameter") 

- During the `fit()` process, the estimator tries to learn the patterns or relationships within the data, and estimates the needed parameters or coefficients to perform its required task.



Examples:
- `StandardScaler`: Standardize features by removing the mean and scaling to unit variance.
- `KMeans`: Unsupervised ML model which aims to partition your dataset into clusters with common patterns.

#### What are transformers, including reference examples? 

- *Transformers* are estimators that can transform a dataset into a new one, typically by applying some mathematical operation or algorithm - this is often, but not always, used to prepare data for modelling. Transformers are often used in preprocessing, feature selection, and feature extraction tasks.



- After applying the `.fit()` method, the transformation is performed by the `transform()` method with the training dataset as the input parameter, returning the transformed dataset. 

> All transformers also have a method called `fit_transform()` - which is the equivalent to calling `fit()` and then `transform()` - that can be more convenient and efficient for transforming the training data.



- Examples:
    - `SimpleImpute`: Impute missing values from a dataset with simple strategies ( mean, median, mode etc.).
    - `OneHotEncoder`: Encode categorical features numerically using a one-hot encoding scheme.
    - `MinMaxScaler`: Transform features by scaling each feature to a given range.

#### What are predictors, including reference examples?

- *Predictors* are estimators that are further capable of making predictions. They are mostly used to solve supervised ML problems.

- A predictor object has a `predict()` method that takes a sample dataset as a parameter and returns its corresponding predictions. It also has a `score()` method to check the performance of the model given a test dataset.



- Examples:

    - `LogisticRegression`: Supervised ML model used for classification problems.
    - `KNeighborsClassifier`: Another supervised ML model used for classification problems.
    - `LinearRegression`: Supervised ML model which aims to explain the relationship between variables.
    
> Note that, for classification problems, instead of predicting the target label directly, you can also use`.predict_proba()` to display the probability for each class for a given sample.

#### What are the main sublibraries of `sklearn`? 

Sklearn is a large API that can perform a diverse range of tasks. It is usefully divided into sublibraries, including the following:

- `sklearn.datasets`: Includes utilities to load built-in datasets.
- `sklearn.impute`:  Includes transformers for missing value imputation.
- `sklearn.preprocessing`: Includes scaling, centering, normalization, and binarization methods.
- `sklear.model_selection`: Includes useful functions for model selection, including train-test split and cross-validation
- `sklearn.linear_model`: Implements a variety of linear models.
- `sklearn.cluster`: Gathers popular unsupervised clustering algorithms.
- `sklearn.neighbors`: Implements the k-nearest neighbors algorithm.
- `sklearn.svm`: Includes Support Vector Machine algorithms.
- `sklearn.tree`: Includes decision tree-based models for classification and regression.
- `sklearn.ensemble`: Includes ensemble-based methods for classification, regression and anomaly detection.
- `sklearn.metrics`: Includes score functions and performance metrics for evaluating ML models.

#### What is feature engineering?

Feature engineering is a typical data preprocessing step that involves using domain knowledge to select, manipulate or/and transform your data set into new features.

The goal of feature engineering is to extract useful features from raw data that can help improve the performance of a machine learning model.

#### Feature Encoding

Encodes categorical variables as numerical ones so that they can be used by machine learning algorithms.

#### One Hot Encoding

`One-hot encoding` generates dummy variables which separate a categorical variable with n categories into n different variables, where each datapoint takes a binary, 0 or 1, value based on whether that datapoint is in the given category or not in the original variable.

### How can `sklearn` be used for preprocessing?

#### What types of preprocessing can be performed in `sklearn`?

- `Encoding`: Transforming non-numerical data into an equivalent numerical form.
- `Feature Selection`: Dropping features and keeping only a selection that we deem useful for our model. 
- `Scaling`: Bringing all of the numerical features into a common range.
- `Imputation`: Replacing missing data with statistical estimates of the missing values
- `Discretization`: Dividing a continuous feature into a pre-specified number of categories.
- `Dataset Balancing`: Creating new datapoints to balance your target values

### Deep Dive: How Feature Scaling Works
Before we build our final pipeline, let's look at exactly what `StandardScaler` does. 

Look at our numerical features. `Visibility_10m` ranges up to 2000, while `Solar_Radiation_MJm2` maxes out around 3.5. If we don't scale these, the machine learning algorithm might mistakenly assume Visibility is thousands of times more important than Solar Radiation simply because the numbers are bigger.

Let's apply the `StandardScaler` to just a few of these columns to see how it forces them all onto the exact same playing field (a mean of 0 and a standard deviation of 1).

```python
# for this dataset , some numerical features have vastly different scales. For example, Visibility_10m can range from 0 to 20000, while Wind_speed_ms typically ranges from 0 to 10. This can cause issues for many machine learning algorithms that are sensitive to the scale of the data, such as linear regression or k-nearest neighbors. To address this, we will apply standard scaling to these numerical features.
# First we import the StandardScaler from sklearn.preprocessing, then we will select the numerical columns that have vastly different scales, fit the scaler to those columns, and transform them. Finally, we will convert the scaled data back to a DataFrame for easier interpretation.
from sklearn.preprocessing import StandardScaler

# 1. Select a few numerical columns that have vastly different scales
numeric_cols = ['visibility_10m', 'solar_radiation_mjm2', 'wind_speed_ms']

# Look at the original data (Notice how large Visibility is compared to Wind Speed)
print("--- ORIGINAL NUMERICAL DATA ---")
print(X_train[numeric_cols].head())
print("\n")
```

```python

# 2. Initialize the StandardScaler
scaler = StandardScaler() 

# this will standardise the features by removing the mean and scaling to unit variance
# you can use other scalers like MinMaxScaler, RobustScaler, etc. depending on the distribution of your data and the requirements of your model
# For example, MinMaxScaler scales the data to a fixed range (usually 0 to 1), while RobustScaler is robust to outliers and scales the data according to the interquartile range. 
# The choice of scaler can depend on the specific characteristics of your dataset and the machine learning algorithm you plan to use.
# For this dataset, since we have features with vastly different scales and we want to ensure that they are on the same scale for modeling, StandardScaler is a good choice as it will center the data around 0 and scale it to have a standard deviation of 1.

```

Now we can apply similar workflow to convert our categorical features to numerical

```python
# Import OneHotEncoder for encoding categorical variables
from sklearn.preprocessing import OneHotEncoder

# Define our three categorical columns
cat_cols = ['seasons', 'holiday', 'functioning_day']


# 1. Look at the original 'Seasons' data (first 5 rows)
print("--- ORIGINAL DATA ---")
print(X_train[cat_cols].head())
print("\n")
```

#### Step 2
* sparse_output=False ensures we get a readable numpy array back instead of a compressed sparse matrix
* drop='first' drops 'Autumn' to prevent perfect multicollinearity

```python

# 2. Initialize the OneHotEncoder

# The OneHotEncoder will create new binary columns for each category in the categorical features. For example, for the 'Seasons' column, it will create separate columns for 'Spring', 'Summer', 'Fall', and 'Winter', with a value of 1 indicating the presence of that season and 0 otherwise. This allows us to convert categorical variables into a format that can be provided to machine learning algorithms to improve predictions.
encoder = OneHotEncoder(sparse_output=False, drop='first') 
# Why do we need to set drop='first' in the OneHotEncoder?
# Setting drop='first' in the OneHotEncoder is a common practice to avoid the dummy variable trap, which occurs when one of the dummy variables can be perfectly predicted from the others. By
```

##### Step 3

* This will create new columns for each category in the `seasons` column (except for the dropped 'Autumn' category), as well as for `'Holiday'` and `'Functioning_Day'`. Each new column will have a value of 1 if the original row belonged to that category, and 0 otherwise. This process allows us to convert categorical variables into a format that can be used by machine learning algorithms.
* For example, if the original `seasons` column had categories 'Spring', 'Summer', 'Autumn', and 'Winter', after one-hot encoding with `drop='first'`, we would get new columns 'Seasons_Spring', 'Seasons_Summer', and 'Seasons_Winter'. If a row had 'Spring' in the original 'Seasons' column, it would have a 1 in 'Seasons_Spring' and 0 in the other two new columns. If it had 'Autumn', it would have 0 in all three new columns since 'Autumn' is the dropped category.
* Similarly, for the `'Holiday'` column with categories 'No Holiday' and 'Holiday', we would get a new column 'Holiday_Holiday' that would be 1 if the original value was 'Holiday' and 0 if it was 'No Holiday'. 
* For the `'Functioning_Day'` column with categories 'Yes' and 'No', we would get a new column 'Functioning_Day_Yes' that would be 1 if the original value was 'Yes' and 0 if it was 'No'.

```python

# 3. Fit and transform just the Seasons column
encoded = encoder.fit_transform(X_train[cat_cols]) # Fill in the blank to fit and transform the all categorical columns defined in cat_cols
```

```python

# 4. Extract the new column names the encoder generated
new_column_names = encoder.get_feature_names_out(cat_cols)
# This will give us the names of the new columns that were created by the one-hot encoding process. The get_feature_names_out method takes the original column names as input and returns the corresponding new column names in the format 'original_column_category'. For example, if we had a 'Seasons' column with categories 'Spring', 'Summer', and 'Winter' (after dropping 'Autumn'), we would get new column names like 'Seasons_Spring', 'Seasons_Summer', and 'Seasons_Winter'. This allows us to easily identify which new columns correspond to which original categories.
# For the 'Holiday' column with categories 'No Holiday' and 'Holiday', we would get a new column name 'Holiday_Holiday'. For the 'Functioning_Day' column with categories 'Yes' and 'No', we would get a new column name 'Functioning_Day_Yes'. These new column names help us understand which original category each new binary column represents after one-hot encoding.

new_column_names
```

```python

# 5. Convert it back to a DataFrame so we can read it easily

encoded_df = pd.DataFrame(encoded, columns=new_column_names, index=X_train.index)

print("--- TRANSFORMED DATA (ONE-HOT ENCODED) ---")
print(encoded_df.head())



encoded_df.shape
```

* We drop the original categorical and numerical columns from `X_train` and concatenate the new encoded and scaled DataFrames to create a fully processed training set that is ready for modeling.
* By doing this, we ensure that our training data now contains only the processed features (one-hot encoded categorical features and standard scaled numerical features) that can be effectively used by machine learning algorithms to learn patterns and make predictions.
* This step is crucial because most machine learning algorithms cannot handle raw categorical data or features with vastly different scales, so we need to preprocess the data to make it suitable for modeling.
* After this step, `X_train_processed` will contain the new one-hot encoded columns for the categorical features and the standard scaled columns for the numerical features, while the original unprocessed columns will be dropped. This processed DataFrame can then be used to train machine learning models effectively.


```python
# merge the new one-hot encoded columns and standard scaled columns back to the original DataFrame
# We drop the original categorical and numerical columns from X_train and concatenate the new encoded and scaled DataFrames to create a fully processed training set that is ready for modeling.
# By doing this, we ensure that our training data now contains only the processed features (one-hot encoded categorical features and standard scaled numerical features) that can be effectively used by machine learning algorithms to learn patterns and make predictions.
# This step is crucial because most machine learning algorithms cannot handle raw categorical data or features with vastly different scales, so we need to preprocess the data to make it suitable for modeling.
# After this step, X_train_processed will contain the new one-hot encoded columns for the categorical features and the standard scaled columns for the numerical features, while the original unprocessed columns will be dropped. This processed DataFrame can then be used to train machine learning models effectively.
X_train_processed = pd.concat([X_train.drop(cat_cols + numeric_cols, axis=1), encoded_df, scaled_df], axis=1)
print(f"Features Shape after preprocessing: {X_train_processed.shape}")
X_train_processed.head()
```

### Model evaluation using `sklearn`

#### What is model evaluation?

After training, the performance of the model must be evaluated against a `scoring metric` to ensure it can generalise well when presented with new and unseen data.

To do this, we set aside a portion of our dataset that will be completely unseen during model training - we call this the `test set`. 

You can choose how much data to set aside for testing; however, you want to use as much data as you can for model training to yield more accurate predictions, especially if you are working with a smaller dataset, we recommend between 10-30% of data set aside for testing and the rest used for training and validation.

We then calculate the performance metric of our trained model on the test set and evaluate the results. We can evaluate our models using a number of different metrics depending on the problem at hand.
```python
import pandas as pd
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

# 1. Assume X_train_scaled is your training data that has already been passed 
# through a StandardScaler (and OneHotEncoder if applicable).
# Assume y_train is your target variable.

# 2. Initialize the Base Estimator
# RFE needs a model that can assign weights/coefficients to features. Linear Regression is perfect for this.
estimator = LinearRegression()

```

```python

# 3. Initialize the RFE Selector
# Let's say we want to narrow our dataset down to the Top 5 most important features.
# step=1 means it will drop the 1 least important feature at a time, recalculate, and repeat.
selector = RFE(estimator=estimator, n_features_to_select=5, step=1)

```

```python

# 4. Fit the RFE Selector to the SCALED training data
print("Running RFE...")
selector.fit(X_train_processed, y_train)

```

```python
# 5. Transform the dataset
# This creates a new version of your training data containing ONLY the Top 5 features!
X_train_selected = selector.transform(X_train_processed)
# --- Reviewing the Results ---
```

```python
# If you have a list of your column names (e.g., feature_names), you can easily see which ones won:
results = pd.DataFrame({
    'Feature': X_train_processed.columns,                # Your list of column names
    'Keep_Feature': selector.support_,       # True if it made the Top 5, False if dropped
    'Elimination_Rank': selector.ranking_    # 1 means Top 5. 2 means dropped last, etc.
})

# Sort by rank so the best features are at the top
results = results.sort_values(by='Elimination_Rank')

print("\n--- RFE Feature Selection Results ---")
print(results.head(10)) # Look at the top 10 rows


```

### Optional stretch - how could I put the data preprocessing and model fitting together in `sklearn` ?

The answer: Use `pipeline`

#### What is a `pipeline`?

- The `pipeline` object is a sklearn class used to chain multiple ML operations (preprocessing, training, predictions) into one object. A pipeline encapsulates different estimators in a single one, applying them sequentially once called. 
- The pipeline object takes a list of `("name", "estimator")` tuples as input, where `name` is a string used to label the step, and `transform` is an estimator object.
- Depending on what estimators you have added to the list, you can call `.fit()`, `.transform()` and `.predict()` on your pipeline, execute these methods tasks on each estimator.



![image.png](https://vitalflux.com/wp-content/uploads/2022/09/Machine-learning-pipeline-Sklearn.png)

- [source](https://vitalflux.com/wp-content/uploads/2022/09/Machine-learning-pipeline-Sklearn.png)

#### What are the benefits of using `pipeline`?

- Make your code easier to read and understand. 
- Make your workflow more convenient as you only have to call fit and predict once on your data to fit a whole sequence of estimators.
- Enforce the execution and order of steps in your project, ensuring that the same sequence of steps is applied to both train and test sets.
- Able to grid search over parameters of all estimators in the pipeline at once.

```python
# Fit the above model using pipeline and RFE together to streamline the process of feature selection and model fitting. This way, you can easily integrate the feature selection step into your modeling workflow and ensure that it is applied consistently during both training and testing phases.

# You can use the same evaluation metrics (MAE, MSE, R^2) to evaluate the model on the test set after preprocessing the test data in the same way as the training data. This will give you an idea of how well your model generalizes to unseen data.

# --- IGNORE ---

# Create a pipeline that includes both the preprocessing steps and the RFE feature selection
preprocessor = ColumnTransformer(transformers=[
    ('num', StandardScaler(), numeric_cols),
    ('cat', OneHotEncoder(sparse_output=False, drop='first'), cat_cols)
]) 
# Create a pipeline that includes the preprocessor, RFE feature selection, and the linear regression model
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('feature_selection', RFE(estimator=LinearRegression(), n_features_to_select=5, step=1)),
    ('model', LinearRegression())
])
pipeline.fit(X_train, y_train)

# Make predictions and evaluate the model on the test set
y_test_pred_pipeline = pipeline.predict(X_test)
test_mae_pipeline = mean_absolute_error(y_test, y_test_pred_pipeline)
test_mse_pipeline = mean_squared_error(y_test, y_test_pred_pipeline)
test_r2_pipeline = r2_score(y_test, y_test_pred_pipeline)
print(f"Test MAE with Pipeline: {test_mae_pipeline:.2f}")
print(f"Test MSE with Pipeline: {test_mse_pipeline:.2f}")
print(f"Test R^2 with Pipeline: {test_r2_pipeline:.4f}")

# Display the selected features from the pipeline
# We can access the RFE step in the pipeline to see which features were selected
rfe_step = pipeline.named_steps['feature_selection']
selected_features = rfe_step.get_support(indices=True)
feature_names = preprocessor.get_feature_names_out()
selected_feature_names = feature_names[selected_features]
print("\nSelected Features from Pipeline:")
print(selected_feature_names)

```

#### Optional stretch 2 - How can ML be described as computational statistical inference?

Statistical inference is the process of using data and statistical models to draw conclusions about a population  based on a sample of the data.

Machine learning can be considered as a form of statistical inference analysis because it often involves using sample data to train a statistical model, and then using the learned model to make predictions. The predictions are based on the probabilities and relationships learned from the data and are therefore considered to be inferences.

Lastly, the use of algorithms to carry out this whole prediction process makes ML a form of computational statistical inference.

As an example, imagine that a credit card company wants to detect fraudulent transactions made by its customers. They collect a large dataset of past transactions and label them as either "fraudulent" or "not fraudulent". They then use this data to train a machine learning model that predicts the likelihood of a transaction being fraudulent based on features such as the transaction amount, location, time of day, etc.

The algorithm here would estimate the coefficients of the model that best separate the transactions which are considered as fraud from those which are not. After the model is trained, you can use it to make predictions for new transactions by inputting the transaction amount and location, and computing the predicted probability of "fraudulent".

#### What is the gap between explanatory and associative modelling?

`Explanatory` and `associative` modeling are two different approaches to building statistical models. The gap between the two lies in their goals and the type of information they aim to uncover from the data.

In `associative modeling`, we aim to understand the relationship and find patterns between our data features (inputs) and the target (output) to make predictions for new instances.

On the other hand, `explanatory modeling` seeks to discover correlations and relationships within data, but not necessarily to explain them.

Consider the example of predicting housing prices. In associate modeling, we would try to understand the underlying causes of housing prices, such as the location, the size, the age of the house, and other factors. We would then build a model that includes these variables and use the model to make predictions about housing prices. 

While in explanatory modeling, we would simply look for patterns in the data without trying to understand the underlying causes. For example, we might identify a strong correlation between housing prices and the proximity to a park and we might even try to prove that there is a statistically significant relation between these variables.
