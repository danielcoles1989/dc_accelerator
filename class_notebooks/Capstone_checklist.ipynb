# Capstone Project: Checklist

> **Note:** Use this checklist to guide your project

---

### 1. Project Scoping & Data Ethics
* [ ] **The Question:** Have I stated a clear, testable hypothesis and my initial expectations?
* [ ] **Relevance:** Did I explain how this data applies to the organization’s specific goals?
* [ ] **Ethics & Privacy:** Have I documented data sharing issues, privacy concerns, or specific guidelines followed?

### 2. Data Validation & Preparation
* [ ] **Cleaning:** Have I performed and documented cleaning tasks (handling nulls, outliers, duplicates)?
* [ ] **Justification:** Did I explain *why* I made specific cleaning decisions based on the business context?
* [ ] **Transformations:** Did I explain how variable transformations (scaling, encoding) impact the data and analysis?
* [ ] **Validation:** Is the final dataset confirmed as "analysis-ready" for all variables?

### 3. Exploratory Data Analysis (EDA)
* [ ] **Statistical Depth:** Have I used statistical tools to summarize core features and distributions?
* [ ] **Univariate Visuals:** Do I have at least **two** different visualisations of single variables (e.g., histograms, bar charts)?
* [ ] **Multivariate Visuals:** Do I have at least **one** visualisation showing relationships between 2+ variables (e.g., scatterplots, heatmaps)?
* [ ] **Interpretation:** Does every visualisation have a caption explaining how it supports the narrative?

### 4. Model Fitting & Evaluation
* [ ] **Problem ID:** Did I correctly identify the problem type (Regression, Classification, or Time series forecasting)?
* [ ] **Baseline vs. Comparison:** Have I fitted a simple **Baseline** model and a more complex **Comparison** model?
* [ ] **Selection Justification:** Did I explain why these specific analytical techniques were chosen?
* [ ] **Model Fit:** Did I discuss model fit (overfitting/underfitting) and evaluate the performance of both approaches?

### 5. Business Focus & Metrics
* [ ] **The KPI:** Have I defined a specific Business KPI to measure the model's success against business criteria?
* [ ] **Actionable Insights:** Have I provided at least one clear proposal/recommendation for future action by the company?
* [ ] **Conclusion:** Did I summarize how this work directly addresses the original business goals?

### 6. Code Quality & Narrative
* [ ] **Error-Free:** Does the code run from start to finish without errors?
* [ ] **Efficiency:** Have I used **functions** to avoid duplicating blocks of code?
* [ ] **Readability:** Is the code annotated with comments explaining the *logic* behind the steps?
* [ ] **Storytelling:** Does the project have a clear narrative that a non-specialist reader can follow?

# ❓ Frequently Asked Questions (FAQ)

**Q: Do I need a high-performing model (e.g., 99% accuracy) to pass?** 

**A:** No. Your capstone is evaluated is based on your **process**. You will receive feedback on how you evaluate the model, how you compare it to a baseline, and how you interpret the results—even if the model's predictive power is limited.

**Q: What is the difference between a Baseline and a Comparison model?** 

**A:** A **baseline** is the simplest reasonable approach (e.g., a Logistic Regression or a Decision Tree). The **comparison** model is a different or more complex algorithm (e.g., Random Forest or Gradient Boosting). You must show you understand why one performed differently than the other.

**Q: Why do I need to define a KPI?** 

**A:** In a business environment, "Accuracy" or "MSE" rarely mean anything to stakeholders. A **KPI** (Key Performance Indicator) translates your model's performance into business value (e.g., "Expected reduction in customer churn cost").

**Q: Can I just delete rows with missing data?** 

**A:** Be careful! The rubric rewards "Data Validation." If you remove data, you must justify *why* it was the right choice. Sometimes imputation or flagging missing values is a better approach for the analysis.

**Q: What should be included in the "Presentation" aspect?** 

**A:** Ensure all charts have clear titles, X/Y axis labels, and legends. Avoid "technical jargon" in your written summaries—write as if you are presenting to a Department Manager, not just a fellow Data Scientist.


**Q: Why do I need to justify "Variable Transformations" like scaling or log transforms?**

**A:** Because transformations change how you interpret the results. For example, if you use a **Log Transform** on price, a coefficient in your model no longer represents a dollar-for-dollar change, but a percentage change. You could demonstrate that you have gone above and beyond what has been covered in this Accelerator by showing competence in interpreting the transformation made on features / target variables.

**Q: What if my attempt to "Improve the Model" actually makes the performance worse?**

**A:** That is perfectly fine! In science, a "failed" experiment is still a result. If you tried Hyperparameter Tuning or Feature Selection and the accuracy dropped, document it. Explain *why* you think it happened (e.g., "The model began to overfit the noise in the training set"). The rubric rewards the **evaluation and discussion** of the attempt, not just the final score.

**Q: How do I discuss "Limitations" without making my work look "bad"?**

**A:** Discussing limitations shows high-level critical thinking. It doesn't mean your work is bad; it means you understand the boundaries of your data. For example: "This model was trained on 2023 budget data; its performance might degrade in 2026 due to inflation or changing spending trends." This helps a policy makers know when to trust the model and when to retrain it.

**Q: What does "comprehensible by a reader without specialist knowledge" really mean?**

**A:** Imagine your project is being read by stakeholders who do not have or little domain knowledge. They care about the **"So What?"** rather than the "How." Avoid saying: "The $p$-value was $< 0.05$." Instead, say: "The relationship between budget spend and income disparity is statistically significant, meaning we can be confident that increasing our budget will reduce disparity." Use your text to translate the math into policy initiatives.

### Critical Thinking Checklist
*Before you finalise your project, consider these three points:*

1. **The "So What?" Factor:** Does my conclusion tell a stakeholder exactly what they should do next?
2. **The "Repeatability" Factor:** Could a colleague use my notebook to reproduce these results without asking me for help?
3. **The "Clarity" Factor:** If I removed all the code, would the remaining text and visuals still tell a complete story?
