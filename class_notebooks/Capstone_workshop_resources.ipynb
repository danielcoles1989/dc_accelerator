## The "Ministerial Briefing" Standard
Before we begin, remember the primary goal: The Narrative.
* A professional data science project explains why a methodology was chosen, not just which code was executed.
* The Test: If a Senior Civil Servant or Minister were to read your report without the code, would the remaining text and visualisations provide a clear, evidence-based case for a specific policy or operational change?

## Breakout Session 1: Policy Pitches & Public Value 
Join your assigned room: Time Series, Regression, or Classification.

### Part A: The 2-Minute Pitch 
Present your project scenario to your peers. Cover the following:
1. The Department/Stakeholder: Who is the service owner? (e.g. DWP, NHS, local authority).
2. The Public Sector Challenge: What specific problem are they facing? (e.g. "We are experiencing a backlog in processing applications" or "We need to predict demand for social care").
3. The Policy Hypothesis: What do you expect the data to reveal? (e.g. "I believe regional demographic shifts are the primary driver of service demand").

### Part B: The Public Value Metric 
Work with a partner to translate your technical goal into a Public Value Metric.
* Don't just say: "I want 90% accuracy."
* Do say: "By achieving 90% accuracy in predicting demand, we can reallocate resources to reduce citizen waiting times by 15%, improving overall service satisfaction."
* Challenge your partner: "If your model is deployed, what specific operational change should the Department implement tomorrow morning?"

## Breakout Session 2: "The Roast" (11:00 – 11:30)
Evaluating Evidence-Based Reports

In your groups, you will critique a "mediocre" sample project. Use the self-assessment rubric to identify three specific areas where it fails to meet the standard. Look for:
* Silent Visualisations: Charts that lack a clear explanation of how they influence policy.
* Lack of Transparency: Removing data or handling outliers without a documented, ethical rationale.
* Technical Obscurity: Using data science jargon without "translating" it for non-specialist policy leads.

## Individual Sprint Roadmap 
Use this as your checklist to maintain momentum.

### Sprint 1: Data Audit & Validation 
* [ ] Integrity Check: Run `df.info()`. Are your data types correct?
* [ ] Ethical Cleaning: Handle nulls and outliers.
* Pro Tip: If you are not using PyCaret, use Pipelines now. This ensures your work is reproducible and auditable—a key requirement for government work.

### Sprint 2: Exploratory Storytelling 
* [ ] Univariate Analysis: Show the distribution of your target (e.g. number of service users).
* [ ] Relationship Analysis: Show how different features correlate (e.g. socioeconomic factors vs. outcomes).
* Reflection: For every chart, write a "Policy Insight" sentence. (e.g. "This trend reveals a 10% spike in demand during winter months, suggesting a need for seasonal staffing increases.")

### Sprint 3: The Analytical Engine 
* [ ] Baseline: Run your initial models to establish a "status quo" performance.
* [ ] Selection: Select the most appropriate model.
* Reflection: Why this model? In the public sector, interpretability (knowing why a model made a choice) is often more important than a 1% gain in accuracy.

### Sprint 4: Finalisation & Recommendations 
* [ ] Optimisation: Finalise your parameters.
* [ ] Strategic Conclusion: Summarise your findings.
* Reflection: Link your final results directly back to the Public Value Metric you defined.

## End-of-project reflection
Before finishing, check the following:

1. Reproducibility: Does the notebook run from top-to-bottom without errors?
2. Comparison: Do you clearly compare a Baseline approach with your Finalised model?
3. Clarity: Is there a clear, non-technical recommendation written for a Policy Lead?
4. Public Value: Is your model's success expressed in real-world outcomes (Citizens helped, £ saved, or Time reduced)?
