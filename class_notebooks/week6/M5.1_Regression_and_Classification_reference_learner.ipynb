### Key Terms

|   Term | Definition         | 
|:-------|:--------------|
| regression | a statistical modeling technique that examines the relationship between a continuous dependent variable and one or more independent variables.|
| simple linear regression | regression with a single dependent variable and a single independent variable|
| residual | the difference between the observed value of the dependent variable and the predicted value from the regression model|
| linearity | the assumption that the relationship between the dependent variable and the independent|
| homoscedascity | the assumption that the variability of the residuals is constant across all levels of the independent variables|
| normality |the assumption that the residuals follow a normal distribution, allowing for valid statistical inference and reliable estimation of model parameters|
| independence |the assumption that the residuals or errors of the regression model are not influenced by or dependent on each other, ensuring that the observations are not correlated|
| intercept |  the estimated value of the dependent variable when all independent variables are zero|
| coefficient | the slope associated with an independent variable, indicating the change in the dependent variable for a one-unit change in that independent variable while holding other variables constant|
| parameter | intercept or coefficient|
| p-value | the p-value associated with a coefficient represents the probability of observing a coefficient as extreme or more extreme than the one estimated, assuming the null hypothesis that the coefficient is zero is true, providing a measure of the statistical significance of the coefficient|
| explanation | the use of regression analysis to understand and explain the relationships between variables, uncovering the factors that influence a dependent variable and quantifying their impact|
| target | the dependent variable or outcome variable that is being predicted or modeled based on the independent variables|
| metric | a quantitative measure used to evaluate the performance or goodness of fit of a regression model, assessing the accuracy or predictive capability of the model in estimating the target variable.|
| binary classification | a classification problem with a class variable that can only take 2 possible values|
| multiclass classification |a classification problem with a class variable that can take 3 or more possible values|
| multilabel clasification | a classification problem with multiple class variables|
| logistic regression | a binary classification algorithm that extends the linear regression formalism to estimate the probability of an event occurring by fitting a logistic function to the input features|
| decision trees | a non-linear classification algorithm that classifies by splitting data successively at a series of nodes|
| cost function | a mathematical measure used to quantify the difference between the predicted values of a model and the actual values in order to optimize and assess the performance of the model during training|
| gradient descent| an iterative optimization algorithm used in machine learning to minimize the loss or cost function of a model by adjusting the model's parameters in the direction of steepest descent of the gradient|
| recursive partitioning | a family of decision algorithms that find the "best" tree by maximising purity at each node|
| impurity metrics | a mathematical function that measures the impurity of a dataset e.g. entropy, gini|
| cross-validation | a technique used to assess the performance and generalizability of a predictive model. It involves dividing a dataset into multiple subsets or "folds," with each fold used as both a training set and a validation set in a series of model iterations|
| grid search | a hyperparameter optimization technique that systematically searches through a predefined set of hyperparameter values to find the optimal combination for a machine learning model|
| confusion matrix |s a tabular representation that summarizes the performance of a classification model by showing the counts of true positive, true negative, false positive, and false negative predictions|
| accuracy | a performance metric for classification models that measures the proportion of correct predictions, calculated as the ratio of the number of correctly classified instances to the total number of instances|
| precision | a performance metric for classification models that measures the proportion of correctly predicted positive instances out of all instances predicted as positive, indicating the model's ability to avoid false positive errors|
| recall | a performance metric for classification models that measures the proportion of correctly predicted positive instances out of all actual positive instances, indicating the model's ability to capture true positive cases|
| f1 score | a single metric that combines precision and recall, providing a balanced measure of a classification model's performance by calculating the harmonic mean of the two metrics|
